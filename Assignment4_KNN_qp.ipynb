# Text Classification using KNN

```python
import numpy as np
import scipy.sparse
```

```python
train_X, test_X = scipy.sparse.load_npz('training_feats.npz').toarray(), scipy.sparse.load_npz('test_feats.npz').toarray()
train_y, test_y = np.load('training_labels.npy', allow_pickle=True), np.load('test_labels.npy', allow_pickle=True)
```

```python
def getAccuracy(test_y, pred):
    cnt = 0
    for i in range(len(pred)):
        if test_y[i]==pred[i]:
            cnt += 1
    acc = float(cnt / len(pred))
    return acc
```

```python
def KNN_Classifier_Euclidean(test_X, train_X, train_y, k_estimators):
    # Euclidean Distance ((x00-x01)**2 + (x10-x11)**2 + ...)**(1/2)
    minus_res = np.tile(test_X, (train_X.shape[0],1)) - train_X # minus operation in one time 
    square_minus_res = minus_res**2 # square each number of minus_res
    square_distances = square_minus_res.sum(axis=1) # sum up
    distances = square_distances**0.5 #sqrt each number of square_distances
    ascend_distance_idx = distances.argsort() # ascending sort by distance and return their index

    # neighbors vote and return the most frequent 
    label_count={}
    for i in range(k_estimators):
        vote_label = train_y[ascend_distance_idx[i]]
        label_count[vote_label] = label_count.get(vote_label,0) + 1
    return max(label_count,key=label_count.get)
```

```python
res_Euclidean = []
for i in range(len(test_X)):
    if i % 100 == 0:
        print(i)
    pred = KNN_Classifier_Euclidean(test_X[i],train_X,train_y,20)
    res_Euclidean.append(pred)
res_Euclidean
```

```python
acc_Euclidean = getAccuracy(test_y, res_Euclidean)
acc_Euclidean
```
